## Penanganan Nilai Hilang (Missing Values) dan menghapus kolom yang tidak terpakai


```python
data_airlines = data_airlines.drop(columns=['Unnamed: 0'])
```

```python
data_airlines.isnull().sum()
```

```python
data_airports = data_airports.drop(columns=['Unnamed: 0','alt','tz','dst','tzone'])
```

```python
data_airports.isnull().sum()
```

```python
data_flights = data_flights.drop(columns=['Unnamed: 0','dep_time','sched_dep_time','arr_time','sched_arr_time','hour','minute','time_hour'])
```

```python
data_flights.isnull().sum()
```

```python
data_flights.dropna(subset=['dep_delay', 'arr_delay', 'air_time', 'tailnum'], inplace=True)
```

```python
data_flights.isnull().sum()
```

## Normalisasi Tipe Data




```python
data_airlines.dtypes
```

```python
data_airports.dtypes
```

```python
data_flights.dtypes
```

```python
data_flights['dep_delay'] = data_flights['dep_delay'].astype(int)
data_flights['arr_delay'] = data_flights['arr_delay'].astype(int)
data_flights['air_time'] = data_flights['air_time'].astype(int)
```

```python
data_flights['full_date'] = pd.to_datetime(data_flights[['year', 'month', 'day']])
```

```python
data_flights.dtypes
```

## Menggabungkan Dataset




```python
flights_merged = pd.merge(data_flights, data_airlines.rename(columns={'name': 'airline_name'}), on='carrier', how='left')
flights_merged
```

```python
airports_origin_renamed = data_airports.rename(columns={
    'faa': 'origin',
    'name': 'origin_airport_name',
    'lat': 'origin_lat',
    'lon': 'origin_lon',
})
flights_merged = pd.merge(flights_merged, airports_origin_renamed, on='origin', how='left')
flights_merged
```

```python
airports_dest_renamed = data_airports.rename(columns={
    'faa': 'dest',
    'name': 'dest_airport_name',
    'lat': 'dest_lat',
    'lon': 'dest_lon',
})
flights_merged = pd.merge(flights_merged, airports_dest_renamed, on='dest', how='left')
flights_merged
```

```python
flights_merged.info()
```

## Penanganan Nilai Hilang (Missing Values) pada data baru, dan menambahkan kolom yang diperlukan

```python
flights_merged.isnull().sum()
```

```python
flights_merged.dropna(subset=['dest_airport_name', 'dest_lat', 'dest_lon'], inplace=True)
```

```python
flights_merged['is_delayed_dep'] = flights_merged['dep_delay'] > 0
flights_merged['is_delayed_arr'] = flights_merged['arr_delay'] > 0
```

```python
flights_merged['route'] = flights_merged['origin_airport_name'] + ' to ' + flights_merged['dest_airport_name']
```

```python
flights_merged.describe()
```

## Menangani Outlier



```python
def cap_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return series.clip(lower=lower_bound, upper=upper_bound)

flights_merged['dep_delay'] = cap_outliers_iqr(flights_merged['dep_delay'])
flights_merged['arr_delay'] = cap_outliers_iqr(flights_merged['arr_delay'])
flights_merged['air_time'] = cap_outliers_iqr(flights_merged['air_time'])
```
